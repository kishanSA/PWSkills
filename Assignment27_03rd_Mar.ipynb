#### Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

**Here are the plots with their definition.**

1. **Scatter Plot:** It displays the relationship between two continuous variables. It can be used to identify patterns, trends, or outliers in the data.


2. **KDE Plot:** A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. It represents the data using a continuous probability density curve in one or more dimensions.


3. **Box Plot:** The seaborn boxplot is a very basic plot Boxplots are used to visualize distributions. Thats very useful when you want to compare data between two groups. Sometimes a boxplot is named a box-and-whisker plot.


4. **Violin Plot:** A violin plot is a statistical representation of numerical data. It is similar to a box plot, with the addition of a rotated kernel density plot on each side.


5. **Count Plot:** A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable. It returns the count of total values for each category using bars.


#### Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. 

Note:  timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
sns.set_style('darkgrid')

fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
sns.lineplot(x='timepoint', y='signal', hue='event', size='region', data= fmri)
plt.xlabel("Time Point")
plt.ylabel("Signal")
plt.title("Time vs Signal for Differenct Regions and Events")
plt.show()
```

#### Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

Note:  pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
sns.set_style('whitegrid')
titanic = sns.load_dataset('titanic')
titanic.head()
```

#### Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset. 

```python
import seaborn as sns
sns.set_style('whitegrid')
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.histplot(x='price', hue='cut', data=diamonds, multiple='stack')
plt.title("Diamond Cutwise Price Distribution")
plt.show()
```

#### Q6: Use the "flights" dataset from seaborn to plot a heatmap

```python
import seaborn as sns

flights = sns.load_dataset('flights')
flights.head()
```

```python
# creating pivot dataframe from flights
pivot_df = flights.pivot(index='year', columns='month', values='passengers')
pivot_df
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.set(font_scale=1.2)
sns.heatmap(data=pivot_df, 
            cmap='Set3',
            fmt='d',
            annot=True,
            linewidths=0.5,
            linecolor='black',
            cbar_kws={'label':'Passenger Counts',
                      'orientation':'horizontal'});
```
