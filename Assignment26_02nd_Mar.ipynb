#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

**Matplotlib** is a plotting aka visualization library for the Python programming language and its numerical mathematics extension of NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK. It has wide range of graph to create variety of visualization. Matplotlib makes easy things easy and hard things possible.

**Matplotlib is used for:**

- Create publication quality plots.
- Make interactive figures that can zoom, pan, update. Customize visual style and layout.
- Export to many file formats.
- Embed in JupyterLab and Graphical User Interfaces.
- Use a rich array of third-party packages built on Matplotlib.
- Visualize machine learning models.
- Custom visualization.

The `matplotlib.pyplot` module is a collectionof functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

**Plots that can be plotted using Pyplot module:**

**1. Scatter Plot:** A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for two different numeric variables.

**2. Line Plot:** A Line plot can be defined as a graph that displays data as points or check marks above a number line, showing the frequency of each value.

**3. Bar Plot:** A bar plot or bar graph is a graphical representation of data in which we can highlight the category with particular shapes like a rectangle.

**4. Histogram:** A histogram is a chart that plots the distribution of a numeric variable's values as a series of bars.

**5. Image Plot:** Draws an image plot with a legend strip for the color scale based on either a regular grid or a grid of quadrilaterals

#### Note: Also add title, xlabel, and ylabel to the plot.

A **scatter plot** (aka scatter chart, scatter graph) is a type of plot that displays dots to represent values for two different numeric variables relationship.

````python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
````

````python
plt.figure(figsize=(10,8))
plt.scatter(x, y)
plt.title("Scatter plot x vs y")
plt.xlabel("X axis values")
plt.ylabel("Y axis values")
plt.show()
````

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


The `subplot()` function creates a figure and a grid of plots in a single call, while providing reasonable control over how the individual plots are created. It creates grid of rows and columns and create plot in particular cell.

````python
import matplotlib.pyplot as plt
import numpy as np

# For line 1:
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

# For line 2:
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

# For line 3:
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

# For line 4:
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
````

````python
fig, axs = plt.subplots(2, 2, figsize=(10,10))

# Plot for line 1
axs[0,0].plot(x1, y1)
axs[0,0].set_title("Axis [0,0] - Line 1")

# Plot for line 2
axs[0,1].plot(x2, y2, 'tab:orange')
axs[0,1].set_title("Axis [0,1] - Line 2")

# Plot for line 3
axs[1,0].plot(x3, y3, 'tab:green')
axs[1,0].set_title("Axis [1,0] - Line 3")

# Plot for line 4
axs[1,1].plot(x4, y4, 'tab:red')
axs[1,1].set_title("Axis [1,1] - Line 4")

# setting label for axes
for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
    
plt.show()
````

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

A `bar plot` or bar graph is a type of data visualization used to represent data in the form of rectangular bars. The height of each bar represents the value of a data point, and the width of each bar represents the category of the data. The Matplotlib library in Python is widely used to create bar plots.

**Reasons for using bar plots:**
- Bar graphs/charts provide a visual presentation of categorical data. Categorical data is a grouping of data into discrete groups, such as months of the year, age group, shoe sizes, and animals. 

- Bar graphs can also be used for more complex comparisons of data with **grouped (or "clustered")** bar charts, and **stacked** bar charts.
    - In **grouped (clustered)** bar charts, for each categorical group there are two or more bars color-coded to represent a particular grouping. For example, a business owner with two stores might make a grouped bar chart with different colored bars to represent each store: the horizontal axis would show the months of the year and the vertical axis would show revenue.
    - Alternatively, a **stacked** bar chart stacks bars on top of each other so that the height of the resulting stack shows the combined result. Stacked bar charts are not suited to data sets having both positive and negative values.

````python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
````

````python
fig, axs = plt.subplots(1, 2, figsize=(15,5))

axs[0].bar(x=company, height=profit)
axs[0].set(xlabel="Company Name", ylabel="Profit")

axs[1].barh(y=company, width=profit, color="orange")
axs[1].set(ylabel="Company Name",xlabel="Profit")
plt.show()
````

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box and whisker plot—also called a **box plot**—displays the five-number summary of a set of data. The five-number summary is the minimum, first quartile, median, third quartile, and maximum. It consists of rectangular box.

Box plot is used to visualize the distribution of dataset. It can tell you about your outliers and what their values are.
Use box and whisker plots when you have multiple data sets from independent sources that are related to each other in some way. Examples include: **Test scores between schools or classrooms.**

Box plot is widely used in data analysis, statistical modeling, and data visualization.

````python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
````

````python
dataset = [box1, box2]
fig, ax = plt.subplots()

ax.boxplot(dataset)
ax.set_xticklabels(['Box 1','Box 2'])
ax.set_ylabel("Box Values")
ax.set_title("Box Plot")
plt.show()
````
