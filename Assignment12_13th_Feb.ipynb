#### Q1. Explain why we have to use the Exception class while creating a Custom Exception.

Note: Here Exception class refers to the base class for all the exceptions.

The Exception class is the base class which contains all the built-in errors in it.

Mostly all the errors occurs in code can be handle by Exception class. But there are times, when you need to provide more context in exceptions to deal with specific requirements. You can show your own message instead of built in error message so user can better understand the error/exception.

One should use the Custom Exception class for the following reasons:
* To improve readability of your code.
* To enhance reusability of fetures.
* To provide custome messages/instructions to user for specific use cases.

#### Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

`LookupError` class is the base class for the exceptions that are raised when a key or index used on a mapping or sequence is invalid: `IndexError`, `KeyError`

```python
# KeyError example

dict1 = {1:'one', 2:'two', 3:'three'}

print(dict1[1])
print(dict1[5]) # this will raise KeyError
```

```python
# IndexError example

lst =[1, 2, 3, 4, 5]

print(lst[0])

print(lst[5]) # this will raised IndexError
```

#### Q6. List down some best practices for exception handling in python.

Some of the best practices for exception handling are as below:

**1. Don't catch general Exception**
- It's bad form to catch the genratl `Exception` class. This will catch every type of exception that subclasses the Exception class, which is almost all of them. You may have errors that you don't care about, and don't affect the operation of your program, or maybe you're dealing with a flaky API and want to swallow errors and retry.

**2. Logging your errors with timestamp**
- You can create log file to catch all kind of error occurs during the running of program. So you can go through it later and fix them.

**3. Try to print always valid message.**

**4. Avoid to write a multiple exception handling**

**5. Create proper documentation for the your program**

**6. Cleanup all the resources at the end of program/application**
